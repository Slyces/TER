```python
import requests
import csv
```

```python
url = "http://chart.finance.yahoo.com/table.csv?s=^GSPC&a=1&b=18&c=2017&d=2&e=18&f=2017&g=d&ignore=.csv"
```

```python
with requests.Session() as s:
    download = s.get(url)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    for row in my_list:
        print(row)
```
